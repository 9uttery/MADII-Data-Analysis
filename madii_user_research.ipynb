```python
# 번외) 클러스터 분석
# 1. 일 - 자기돌봄
from sklearn.cluster import KMeans

scaled_df = pd.DataFrame()
scaled_df['workAvg'] = df['workAvg'] # workAvg
scaled_df['scAvg'] = df['scAvg'] # scAvg

model = KMeans(n_clusters=3, random_state=123)
model.fit(scaled_df)
scaled_df['label'] = model.predict(scaled_df)
scaled_df['label'].value_counts()

sns.scatterplot(x= scaled_df['workAvg'], y= scaled_df['scAvg'], hue= scaled_df['label'], s=200)

groupby_df = scaled_df.groupby('label').mean()
scaled_df['label'].value_counts()
```

```python
# 각 그룹 자기돌봄 못하는 이유 추출해서 보기

import matplotlib.pyplot as plt

def whycount(df, ax):
    why = df['scNoWhy'].dropna()
    # 0 = noTime
    # 1 = noEnergy
    # 2 = xKnow
    # 3 = tiresome
    # 4 = noNeed
    # 5 = etc

    why_cnt = [0 for j in range(6)]
    for i in why.index:
        text = why[i]
        text = text.replace('noTime', '0')
        text = text.replace('noEnergy', '1')
        text = text.replace('xKnow', '2')
        text = text.replace('tiresome', '3')
        text = text.replace('noNeed', '4')
        text = text.replace('etc', '5')
        new_text = text.split(", ")
        for j in range(len(new_text)):
            for k in range(6):
                if new_text[j] == str(k):
                    why_cnt[k] += 1
    xlabels = ['noTime', 'noEnergy', 'xKnow', 'tiresome', 'noNeed', 'etc']
    total_cnt = sum(why_cnt)
    if total_cnt > 0:
        why_pct = [cnt/total_cnt for cnt in why_cnt]
    else:
        why_pct = [0 for _ in why_cnt]
        
    ax.pie(why_pct, labels=xlabels, autopct='%1.1f%%')
    ax.set_title('Reasons for No Self-care Action')
    return why_cnt

# 데이터 준비
df_g0 = df[scaled_df['label'] == 0]
df_g1 = df[scaled_df['label'] == 1]
df_g2 = df[scaled_df['label'] == 2]

# Figure와 subplot 생성
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# 각 그룹에 대해 whycount 함수 호출 및 subplot에 그래프 그리기
whycount(df_g0, axs[0])
axs[0].set_title('Group 0')

whycount(df_g1, axs[1])
axs[1].set_title('Group 1')

whycount(df_g2, axs[2])
axs[2].set_title('Group 2')

# 그래프를 보기 좋게 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```
